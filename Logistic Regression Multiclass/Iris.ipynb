```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris as li
```

```python
iris=li()
```

```python
dir(iris)
```

```python
iris.data
```

```python
iris.target
```

```python
from sklearn.model_selection import train_test_split as tss
```

```python
xtrain,xtest,ytrain,ytest=tss(iris.data,iris.target,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression as reg
model=reg()
model
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest),model.score(xtrain,ytrain)
```

```python
y_predicted=model.predict(xtest)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ytest,y_predicted)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
```
