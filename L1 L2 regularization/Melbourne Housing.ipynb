```python
cols= ['Suburb', 'Rooms','Type','Method', 'SellerG', 'Regionname','Propertycount','Distance','CouncilArea','Bedroom2' , 'Bathroom','Car','Landsize','BuildingArea' ,'Price']
melb=melb[cols]
melb.head()
```

```python
cols
```

```python
melb.shape
```

```python
melb.isna().sum()
```

```python
colstofillzero=[ 'Propertycount' ,'Distance','Bedroom2' ,'Bathroom','Car']
```

```python
melb[colstofillzero]=melb[colstofillzero].fillna(0)
```

```python
melb.isna().sum()
```

```python
melb['Landsize']=melb['Landsize'].fillna(melb.Landsize.mean())
```

```python
melb['BuildingArea']=melb['BuildingArea'].fillna(melb.BuildingArea.mean())
```

```python
melb.isna().sum()
```

```python
melb.dropna(inplace=True)
```

```python
melb.isna().sum()
```

```python
melb.shape
```

```python
melb=pd.get_dummies(melb,drop_first=True,dtype=int)
```

```python
melb.head()
```

```python
x=melb.drop('Price',axis=1)
y=melb['Price']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
model=lr().fit(xtrain,ytrain)
```

```python
model
```

```python
model.score(xtest,ytest),model.score(xtrain,ytrain)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg=Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
lasso_reg.fit(xtrain,ytrain)
```

```python
lasso_reg.score(xtest,ytest),lasso_reg.score(xtrain,ytrain)
```
