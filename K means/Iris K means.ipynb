```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
iris=load_iris()
iris
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
pd.DataFrame(iris['data']).head()
```

```python
df=pd.DataFrame(iris['data'],columns=iris.feature_names)
```

```python
df.head()
```

```python
df=df.drop(['sepal length (cm)','sepal width (cm)'],axis='columns')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
yp=km.fit_predict(df)
yp
```

```python
df['cluster']=yp
df.head()
```

```python
df.cluster.unique()
```

```python
km.cluster_centers_
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='purple')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='b')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='g')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='x',color='r',label='centroid')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(df[['petal width (cm)']])
```

```python
df['petal width (cm)']=scaler.transform(df[['petal width (cm)']])
```

```python
scaler.fit(df[['petal length (cm)']])
```

```python
df['petal length (cm)']=scaler.transform(df[['petal length (cm)']])
```

```python
df
```

```python
km=KMeans(n_clusters=3)
yp=km.fit_predict(df)
```

```python
df['cluster']=yp
```

```python
df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='purple')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='b')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='g')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='x',color='r',label='centroid')
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
sse
```

```python
plt.plot(k_rng,sse)
```

```python
for i in range(1,20):
    k_rng=range(1,i)
    sse=[]
    for k in k_rng:
        km=KMeans(n_clusters=k)
        km.fit(df)
        sse.append(km.inertia_)
    plt.plot(k_rng,sse)
```
