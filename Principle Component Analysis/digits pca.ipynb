```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df
```

```python
df.describe()
```

```python
x=df
```

```python
x.shape
```

```python
y=digits.target
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
xscaled=scaler.fit_transform(x)
xscaled
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(xscaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression as lr
```

```python
model=lr()
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain),model.score(xtest,ytest)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(0.95)
```

```python
pca
```

```python
xpca=pca.fit_transform(x)
```

```python
xpca.shape
```

```python
xpca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
xptrain,xptest,yptrain,yptest=tts(xpca,y,test_size=0.2,random_state=30)
```

```python
model=lr(max_iter=1000)
```

```python
model.fit(xptrain,yptrain)
```

```python
model.score(xptrain,yptrain),model.score(xptest,yptest)
```

```python
pca=PCA(n_components=2)
```

```python
xpca=pca.fit_transform(x)
```

```python
xpca.shape
```

```python
pca.explained_variance_ratio_
```

```python
xptrain,xptest,yptrain,yptest=tts(xpca,y,test_size=0.2,random_state=30)
```

```python
model=lr(max_iter=1000)
model.fit(xptrain,yptrain)
model.score(xptrain,yptrain),model.score(xptest,yptest)
```
