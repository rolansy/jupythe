```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
wine=load_wine()
```

```python
dir(wine)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
inputs=pd.DataFrame(wine.data,columns=wine.feature_names)
```

```python
inputs
```

```python
wine.target
```

```python
target=pd.DataFrame(wine.target)
```

```python
target
```

```python
xtrain,xtest,ytrain,ytest=tts(inputs,target,train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain),model.score(xtest,ytest)
```

```python
model.predict(xtest[:10])
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain),model.score(xtest,ytest)
```
