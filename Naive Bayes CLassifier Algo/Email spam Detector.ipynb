```python
import pandas as pd
df=pd.read_csv('spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
dummies=pd.get_dummies(df.Category,dtype=int)
dummies
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(df.Message,df.spam,test_size=0.8)
```

```python
xtrain
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v=CountVectorizer()
```

```python
v
```

```python
xtrain.info
```

```python
x_train_count=v.fit_transform(xtrain.values)
x_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(x_train_count,ytrain)
```

```python
model.score(x_train_count,ytrain)
```

```python
emails=['Hey mohan, can we get together to watch football game tomorrow?' ,'Get upto 20% discount on parking now!! exclusive offer just for you. Do Not miss this reward!']
```

```python
email_count=v.transform(emails)
```

```python
model.predict(email_count)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([('vectorizer',CountVectorizer()),('nb',MultinomialNB())])
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtrain,ytrain),clf.score(xtest,ytest)
```

```python
xtrain,xtest,ytrain,ytest=tts(df.Message,df.spam,test_size=0.9)
```
