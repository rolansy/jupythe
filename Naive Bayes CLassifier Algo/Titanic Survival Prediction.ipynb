```python
import pandas as pd
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
targets=df.Survived
```

```python
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.Sex,dtype=int)
```

```python
dummies
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
```

```python
inputs.head()
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
```

```python
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age.fillna(inputs.Age.mean(),inplace=True)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(inputs,targets,train_size=0.8)
```

```python
len(xtrain),len(ytest)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain),model.score(xtest,ytest)
```

```python
model.predict(xtest[:10])
```

```python
model.predict_proba(xtest[:10])
```
