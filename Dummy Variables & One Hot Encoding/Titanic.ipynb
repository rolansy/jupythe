```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
inputs=df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
inputs.head()
```

```python
target=df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder as le
```

```python
le_sex=le()
```

```python
inputs['Sexn']=le_sex.fit_transform(inputs['Sex'])
```

```python
inputs.head()
```

```python
inputsn=inputs.drop('Sex',axis='columns')
inputsn.head()
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
model=dtc()
model
```

```python
model.fit(inputsn,target)
```

```python
model.score(inputsn,target)
```
