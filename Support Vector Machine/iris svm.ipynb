```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
iris.target_names
```

```python
for i in iris.target_names:
    print(i)
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
import matplotlib.pyplot as plt
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df0.head()
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='g',marker='x')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='b',marker='+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='orange',marker='.')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='g',marker='x')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='b',marker='+')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='orange',marker='.')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X=df.drop(['target','flower_name'],axis='columns')
X
```

```python
y=df['target']
y.head()
```

```python
xtrain,xtest,ytrain,ytest=tts(X,y,train_size=0.8)
```

```python
len(xtrain)
```

```python
from sklearn.svm import SVC
model=SVC(C=5.0,kernel='linear')  #C is regularization, default of which is at C=1.0 and can be increased regularization
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain),model.score(xtest,ytest)
```
