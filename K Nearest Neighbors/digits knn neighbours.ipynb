```python
from sklearn.datasets import load_digits
```

```python
import pandas as pd
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
df=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df
```

```python
digits.target
```

```python
df['target']=digits.target
```

```python
df
```

```python
dfx=[]
for i in digits.target:
    dfx.append(df[df.target==i])
```

```python
dfx
```

```python
import matplotlib.pyplot as plt
```

```python
for i in digits.target:
    plt.scatter(dfx[i]['pixel_0_0'],dfx[i]['pixel_0_1'])
```

```python
from sklearn.model_selection import train_test_split as tss
```

```python
xtrain,xtest,ytrain,ytest=tss(df.drop('target',axis=1),df.target,train_size=0.7)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knc
```

```python
knn=knc(n_neighbors=5)
```

```python
knn
```

```python
knn.fit(xtrain,ytrain)
```

```python
knn.score(xtest,ytest),knn.score(xtrain,ytrain)
```

```python
for i in range(1,10):
    knn=knc(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    print(i,knn.score(xtest,ytest),knn.score(xtrain,ytrain))
```

```python
from sklearn.metrics import confusion_matrix as cmx
```

```python
ypred=knn.predict(xtest)
```

```python
cm=cmx(ytest,ypred)
```

```python
cm
```

```python
import seaborn as sn
```

```python
sn.heatmap(cm,annot=True)
```

```python
for i in range(1,10):
    knn=knc(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    print(i,knn.score(xtest,ytest),knn.score(xtrain,ytrain))
    ypred=knn.predict(xtest)
    cm=cmx(ytest,ypred)
    plt.figure(figsize=(7,5))
    sn.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report as cr
```

```python
print(cr(ytest,ypred))
```
