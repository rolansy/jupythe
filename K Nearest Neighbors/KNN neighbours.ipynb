```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
iris.target
```

```python
df['target']=iris.target
```

```python
df[df.target==1]
```

```python
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='+',color='g')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='*',color='b')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],marker='x',color='r')
plt.legend()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=df.drop(['target','flower_name'],axis=1)
```

```python
y=df.target
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,train_size=0.8,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn
```

```python
knn.fit(xtrain,ytrain)
```

```python
knn.score(xtrain,ytrain),knn.score(xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
ypred=knn.predict(xtest)
```

```python
cm=confusion_matrix(ytest,ypred)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True)
plt.ylabel('Truth')
plt.xlabel('Predicted')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```
